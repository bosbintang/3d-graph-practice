```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

def plot2D(x, y, p):
    fig = plt.figure(figsize=(9, 5), dpi=100)
    ax = fig.gca(projection='3d')
    X, Y = np.meshgrid(x, y)
    surf = ax.plot_surface(X, Y, p[:], rstride=1, cstride=1, cmap='coolwarm',
            linewidth=0, antialiased=False)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.view_init(30, 225)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.view_init(30, 60)
    #fig.colorbar(surf)

# initialization
nx = 10
ny = 10
dx = 1 / nx
dy = 1 / ny
dt = 1/100


u = np.zeros((nx,ny))
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
X, Y = np.meshgrid(x, y)

u = 16*X*Y*(1-X)*(1-Y)

for k in range(10):
     un = u.copy()
     u[1:-1, 1:-1] = ((dy**2 * (un[1:-1,2:] + un[1:-1,0:-2])) + (dx**2 * (un[2:,1:-1] + un[0:-2,1:-1]))) / (2 * (dx**2 + dy**2))

plot2D(x, y, u)
```
